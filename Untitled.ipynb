```python
save_directory = '/Users/tarunbatchu/ComputerVisionSSL/SpacesuitDataSet/final_predictions'
os.makedirs(save_directory, exist_ok=True)  # Create the folder if it doesnâ€™t exist
path = "/Users/tarunbatchu/ComputerVisionSSL/SpacesuitDataSet/ssdata/test/images/090422_MoonYard_029_jpg.rf.09d7352c85b3002858aa73928e205884.jpg"
i = 1
for result in results:
    if len(result.boxes) > 0:  # If there are any detected objects
        best_box = result.boxes.conf.argmax()  # Find the index of the box with max confidence
        best_box_info = result.boxes[best_box]  # Get the best bounding box

        # Extract bounding box coordinates and confidence score
        x1, y1, x2, y2 = best_box_info.xyxy[0]  # Bounding box coordinates
        confidence = best_box_info.conf[0]  # Confidence score
        class_id = int(best_box_info.cls[0])  # Predicted class ID

        print(f"Best Box: Class {class_id}, Confidence: {confidence:.2f}, BBox: ({x1}, {y1}, {x2}, {y2})")

        # Load image and draw best bounding box
        image = cv.imread(result.path)
        cv.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
        cv.putText(image, f"Conf: {confidence:.2f}", (int(x1), int(y1) - 10),
                    cv.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        # Save or show image
        
        # Extract filename with extension
        image_name = os.path.basename(path)
        save_path = os.path.join(save_directory, f'final_prediction_{i}.jpg')
        cv.imwrite(save_path, image)  # Save the image with the best box
#         cv.imshow("Best Prediction", image)  # Show image (for local testing)

#         cv.waitKey(0)
        i += 1
        cv.destroyAllWindows()

```
